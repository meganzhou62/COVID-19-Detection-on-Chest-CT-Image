```python
import pandas as pd
source_path = "/gdrive"
metadata = pd.read_csv(source_path+'/MyDrive/AML_group13/Data/metadata.csv')
```

```python
metadata
```

```python
import seaborn as sns
sns.set_theme(style = "whitegrid")

sns.histplot(
    metadata, x="finding"
).set(title='Distribution of finding label')
```

```python
metadata['finding'].hist()
```

```python
percent_missing = metadata.isnull().sum() * 100 / len(metadata)
missing_value_df = pd.DataFrame({'column_name': metadata.columns,
                                 'percent_missing': percent_missing})
missing_value_df
```

```python
country_finding = metadata.groupby(['country', 'finding']).size().reset_index()
country_finding.columns = ['country', 'finding', 'count']
```

```python
sns.set_theme(style = "whitegrid")
sns.catplot(data = country_finding, kind = "bar", y = "country", x = "count", hue = 'finding', orient ="h")
```

```python
metadata.source.unique()
```

```python
source_finding = metadata.groupby(['source', 'finding']).size().reset_index()
source_finding.columns = ['source', 'finding', 'count']
```

```python
sns.set_theme(style = "whitegrid")
sns.catplot(data = source_finding, kind = "bar", y = "source", x = "count", hue = 'finding', orient ="h")
```

```python
sex_finding = metadata.groupby(['sex', 'finding']).size().reset_index()
sex_finding.columns = ['sex', 'finding', 'count']
```

```python
sns.set_theme(style = "whitegrid")
sns.catplot(data = sex_finding, kind = "bar", y = "sex", x = "count", hue = 'finding', orient ="h")
```

```python
sns.boxplot(x = "finding", y = 'age', data = metadata)
```
